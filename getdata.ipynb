```python
# 清理数据函数
# 如果有2019文件夹，文件移到上级并且名字加上2019；如果只有18、17文件夹，舍去；如果有分工业文件夹，全部移到上级
# 使用哪一年的就用哪一年的函数，其他的不要用
def move_to_root_folder(sourcePath, destPath):
    fileList = os.listdir(sourcePath)
    for filename in fileList:
        filePath = os.path.join(sourcePath, filename)
        if not os.path.isdir(filePath):
            if not filename.startswith('.'):
                shutil.move(filePath, os.path.join(destPath, filename))
            else:
                os.remove(filePath)
        elif os.path.isdir(filePath):
            move_to_root_folder(filePath, destPath)
        else:
            pass

        
def move_to_root_folder_2019(sourcePath, destPath): #把2019文件夹里面的data移到上一级
    fileList = os.listdir(sourcePath)
    for filename in fileList:
        filePath = os.path.join(sourcePath, filename)
        if not os.path.isdir(filePath):
            if not filename.startswith('.'):
                newpath = os.path.join(destPath, filename)
                shutil.move(filePath, newpath)
                os.rename(newpath, '%s%s%s'%(os.path.splitext(newpath)[0],"2019", os.path.splitext(newpath)[1]))
            else:
                os.remove(filePath)
        elif os.path.isdir(filePath):
            move_to_root_folder_2019(filePath, destPath)
        else:
            pass

        

    
def move_to_root_folder_2018(sourcePath, destPath): #把2018文件夹里面的data移到上一级
    fileList = os.listdir(sourcePath)
    for filename in fileList:
        filePath = os.path.join(sourcePath, filename)
        if not os.path.isdir(filePath):
            if not filename.startswith('.'):
                newpath = os.path.join(destPath, filename)
                shutil.move(filePath, newpath)
                os.rename(newpath, '%s%s%s'%(os.path.splitext(newpath)[0],"2018", os.path.splitext(newpath)[1]))
            else:
                os.remove(filePath)
        elif os.path.isdir(filePath):
            move_to_root_folder_2018(filePath, destPath)
        else:
            pass

        

def move_to_root_folder_2017(sourcePath, destPath): #把2017文件夹里面的data移到上一级
    fileList = os.listdir(sourcePath)
    for filename in fileList:
        filePath = os.path.join(sourcePath, filename)
        if not os.path.isdir(filePath):
            if not filename.startswith('.'):
                newpath = os.path.join(destPath, filename)
                shutil.move(filePath, newpath)
                os.rename(newpath, '%s%s%s'%(os.path.splitext(newpath)[0],"2017", os.path.splitext(newpath)[1]))
            else:
                os.remove(filePath)
        elif os.path.isdir(filePath):
            move_to_root_folder_2017(filePath, destPath)
        else:
            pass


        
def move_to_root_folder_2016(sourcePath, destPath): #把2016文件夹里面的data移到上一级
    fileList = os.listdir(sourcePath)
    for filename in fileList:
        filePath = os.path.join(sourcePath, filename)
        if not os.path.isdir(filePath):
            if not filename.startswith('.'):
                newpath = os.path.join(destPath, filename)
                shutil.move(filePath, newpath)
                os.rename(newpath, '%s%s%s'%(os.path.splitext(newpath)[0],"2016", os.path.splitext(newpath)[1]))
            else:
                os.remove(filePath)
        elif os.path.isdir(filePath):
            move_to_root_folder_2016(filePath, destPath)
        else:
            pass


def move_to_root_folder_2015(sourcePath, destPath): #把2015文件夹里面的data移到上一级
    fileList = os.listdir(sourcePath)
    for filename in fileList:
        filePath = os.path.join(sourcePath, filename)
        if not os.path.isdir(filePath):
            if not filename.startswith('.'):
                newpath = os.path.join(destPath, filename)
                shutil.move(filePath, newpath)
                os.rename(newpath, '%s%s%s'%(os.path.splitext(newpath)[0],"2015", os.path.splitext(newpath)[1]))
            else:
                os.remove(filePath)
        elif os.path.isdir(filePath):
            move_to_root_folder_2015(filePath, destPath)
        else:
            pass

       
        
def clearCityFiles2019(cityPath): # 城市数据的路径
    fileList = os.listdir(cityPath)
    if any("201" in s for s in fileList) and not any("2019" in s for s in fileList):
        print("%s has no 2019 data"%cityPath)
    else:
        for filename in fileList:
            filePath = os.path.join(cityPath, filename)
            if os.path.isdir(filePath) and "2019" in filename:
                move_to_root_folder_2019(filePath, cityPath)   
            elif os.path.isdir(filePath) and "201" in filename:
                pass
            elif  os.path.isdir(filePath):
                subfileList = os.listdir(filePath)
                for subfilename in subfileList:
                    subfilePath = os.path.join(filePath, subfilename)
                    if os.path.isdir(subfilePath) and "2019" in subfilename:
                        move_to_root_folder_2019(subfilePath, cityPath)
                    elif os.path.isdir(subfilePath) and "201" in filename:
                        pass
                    elif os.path.isdir(subfilePath):
                        move_to_root_folder(subfilePath, cityPath)    
                    elif not os.path.isdir(subfilePath):
                        shutil.move(subfilePath, os.path.join(cityPath, subfilename))
                    else:
                        pass
            else:
                pass 
        
        
def clearCityFiles2018(cityPath): # 城市数据的路径
    fileList = os.listdir(cityPath)
    if any("201" in s for s in fileList) and not any("2018" in s for s in fileList):
        print("%s has no 2018 data"%cityPath)
    else:
        for filename in fileList:
            filePath = os.path.join(cityPath, filename)
            if os.path.isdir(filePath) and "2018" in filename:
                move_to_root_folder_2018(filePath, cityPath)   
            elif os.path.isdir(filePath) and "201" in filename:
                pass
            elif  os.path.isdir(filePath):
                subfileList = os.listdir(filePath)
                for subfilename in subfileList:
                    subfilePath = os.path.join(filePath, subfilename)
                    if os.path.isdir(subfilePath) and "2018" in subfilename:
                        move_to_root_folder_2018(subfilePath, cityPath)
                    elif os.path.isdir(subfilePath) and "201" in filename:
                        pass
                    elif os.path.isdir(subfilePath):
                        move_to_root_folder(subfilePath, cityPath)    
                    elif not os.path.isdir(subfilePath):
                        shutil.move(subfilePath, os.path.join(cityPath, subfilename))
                    else:
                        pass
            else:
                pass
 

def clearCityFiles2017(cityPath): # 城市数据的路径
    fileList = os.listdir(cityPath)
    if any("201" in s for s in fileList) and not any("2017" in s for s in fileList):
        print("%s has no 2017 data"%cityPath)
    else:
        for filename in fileList:
            filePath = os.path.join(cityPath, filename)
            if os.path.isdir(filePath) and "2017" in filename:
                move_to_root_folder_2017(filePath, cityPath)   
            elif os.path.isdir(filePath) and "201" in filename:
                pass
            elif  os.path.isdir(filePath):
                subfileList = os.listdir(filePath)
                for subfilename in subfileList:
                    subfilePath = os.path.join(filePath, subfilename)
                    if os.path.isdir(subfilePath) and "2017" in subfilename:
                        move_to_root_folder_2017(subfilePath, cityPath)
                    elif os.path.isdir(subfilePath) and "201" in filename:
                        pass
                    elif os.path.isdir(subfilePath):
                        move_to_root_folder(subfilePath, cityPath)    
                    elif not os.path.isdir(subfilePath):
                        shutil.move(subfilePath, os.path.join(cityPath, subfilename))
                    else:
                        pass
            else:
                pass
            

def clearCityFiles2016(cityPath): # 城市数据的路径
    fileList = os.listdir(cityPath)
    if any("201" in s for s in fileList) and not any("2016" in s for s in fileList):
        print("%s has no 2016 data"%cityPath)
    else:
        for filename in fileList:
            filePath = os.path.join(cityPath, filename)
            if os.path.isdir(filePath) and "2016" in filename:
                move_to_root_folder_2016(filePath, cityPath)   
            elif os.path.isdir(filePath) and "201" in filename:
                pass
            elif  os.path.isdir(filePath):
                subfileList = os.listdir(filePath)
                for subfilename in subfileList:
                    subfilePath = os.path.join(filePath, subfilename)
                    if os.path.isdir(subfilePath) and "2016" in subfilename:
                        move_to_root_folder_2016(subfilePath, cityPath)
                    elif os.path.isdir(subfilePath) and "201" in filename:
                        pass
                    elif os.path.isdir(subfilePath):
                        move_to_root_folder(subfilePath, cityPath)    
                    elif not os.path.isdir(subfilePath):
                        shutil.move(subfilePath, os.path.join(cityPath, subfilename))
                    else:
                        pass
            else:
                pass
            

def clearCityFiles2015(cityPath): # 城市数据的路径
    fileList = os.listdir(cityPath)
    if any("201" in s for s in fileList) and not any("2015" in s for s in fileList):
        print("%s has no 2015 data"%cityPath)
    else:
        for filename in fileList:
            filePath = os.path.join(cityPath, filename)
            if os.path.isdir(filePath) and "2015" in filename:
                move_to_root_folder_2015(filePath, cityPath)   
            elif os.path.isdir(filePath) and "201" in filename:
                pass
            elif  os.path.isdir(filePath):
                subfileList = os.listdir(filePath)
                for subfilename in subfileList:
                    subfilePath = os.path.join(filePath, subfilename)
                    if os.path.isdir(subfilePath) and "2015" in subfilename:
                        move_to_root_folder_2015(subfilePath, cityPath)
                    elif os.path.isdir(subfilePath) and "201" in filename:
                        pass
                    elif os.path.isdir(subfilePath):
                        move_to_root_folder(subfilePath, cityPath)    
                    elif not os.path.isdir(subfilePath):
                        shutil.move(subfilePath, os.path.join(cityPath, subfilename))
                    else:
                        pass
            else:
                pass
     
```

```python

# 标题和前两行
def decrypt(filePath, headernumber, indexnumber):#解密
    file=msoffcrypto.OfficeFile(open(filePath,'rb'))
    file.load_key(password='VelvetSweatshop')
    newPath = '%s%s%s'%(os.path.splitext(filePath)[0],"decrypt", os.path.splitext(filePath)[1])
    file.decrypt(open(newPath,'wb'))
    data_excel = pd.read_excel(newPath, header=headernumber, index_col = indexnumber)
    return data_excel

def readfile(filePath, headernumber, indexnumber): #读入
    try:
        table = pd.read_excel(filePath, header=headernumber, index_col = indexnumber)
    except:
        table = decrypt(filePath, headernumber, indexnumber)
    return table
    
        

#def findfiles2016(cityPath):#筛选需要的能源数据、生产数据
#    cityfileList = os.listdir(cityPath)
#    tablenames = []
#    for filename in cityfileList:
#        filePath = os.path.join(cityPath, filename)
#        if not os.path.isdir(filePath):
#            if ("2016" in filename or "2015－" in filename or "2005-" in filename) and (("能源" in filename) or ("总值" in filename) or ("产值" in filename) or ("经济指标" in filename)): 
#                tablenames.append(filename)
#    return tablenames

def findEnergy(cityPath, year):#筛选需要的能源数据1
    cityfileList = os.listdir(cityPath)
    tablenames = []
    for filename in cityfileList:
        filePath = os.path.join(cityPath, filename)
        if (year in filename) and not os.path.isdir(filePath) and not filename.startswith('.') and not "decrypt" in filename and not os.path.splitext(filePath)[1] == ".doc":
            try:
                table = readfile(filePath, 0,0)
                if isinstance(table.index.name, str) and "行业" in table.index.name and "能源" in table.index.name:     
                    if len(table) > 70:
                        tablenames.append([table.index.name, "续表在下", 2*table.shape[1], len(table)])

                    else:
                        tablenames.append([table.index.name, table.shape[1], len(table)])
            except:
                print(filePath)
            else:
                pass
    return tablenames

def findEnergyC(cityPath, year):#筛选需要的能源数据2
    cityfileList = os.listdir(cityPath)
    tablenames = []
    for filename in cityfileList:
        filePath = os.path.join(cityPath, filename)
        if (year in filename) and not os.path.isdir(filePath) and not filename.startswith('.') and not "decrypt" in filename and not os.path.splitext(filePath)[1] == ".doc":
            try:
                table = readfile(filePath, 0,0)
                if isinstance(table.index.name, str):
                    if "行业" in table.index.name and "能耗" in table.index.name: 
                        tablenames.append(table.index.name)
            except: 
                print(filePath)
            else:
                pass
    return tablenames

def findEnergyBalance(cityPath, year):#筛选需要的能源平衡表
    cityfileList = os.listdir(cityPath)
    tablenames = []
    for filename in cityfileList:
        filePath = os.path.join(cityPath, filename)
        if (year in filename) and not os.path.isdir(filePath) and not filename.startswith('.') and not "decrypt" in filename and not os.path.splitext(filePath)[1] == ".doc":
            try:
                table = readfile(filePath, 0,0)
                if isinstance(table.index.name, str):
                    if "能源平衡表" in table.index.name: 
                        tablenames.append([table.index.name, table.shape[1], len(table)])
            except:
                print(filePath)
            else:
                pass
    return tablenames

def findProduct(cityPath, year):#筛选需要的生产数据
    cityfileList = os.listdir(cityPath)
    tablenames = []
    for filename in cityfileList:
        filePath = os.path.join(cityPath, filename)
        if (year in filename) and not os.path.isdir(filePath) and not filename.startswith('.') and not "decrypt" in filename and not os.path.splitext(filePath)[1] == ".doc":
            try:
                table = readfile(filePath, 0,0)
                if isinstance(table.index.name, str):
                    if ("行业" in table.index.name and ("生产" in table.index.name or "指标" in table.index.name or "产值" in table.index.name)) or ("行业" in filename and ("生产" in filename or "指标" in filename or "产值" in filename)): 
                        tablenames.append([table.index.name, table.shape[1], len(table)])
            except: 
                print(filePath)
            else:
                pass
    return tablenames

def findIndustryProduct(cityPath, year):
    cityfileList = os.listdir(cityPath)
    tablenames = []
    for filename in cityfileList:
        filePath = os.path.join(cityPath, filename)
        if "工业" in filename and "产品" in filename and not os.path.isdir(filePath) and not filename.startswith('.') and not "decrypt" in filename and not os.path.splitext(filePath)[1] == ".doc" and not os.path.splitext(filePath)[1] == ".docx":
            try:
                table = readfile(filePath, 0,0)
                if isinstance(table.index.name, str):
                    if "工业" in table.index.name and "产品" in table.index.name and "产量" in table.index.name and (year in table.index or year in table.index.name):
                        tablenames.append([table.index.name, "产量", table.shape[1], len(table)])
                    elif "工业" in table.index.name and "产品" in table.index.name and "存量" in table.index.name and (year in table.index or year in table.index.name):
                        tablenames.append([table.index.name, "存量", table.shape[1], len(table)])       
            except: 
                print(filePath)
            else:
                pass
    return tablenames

```
